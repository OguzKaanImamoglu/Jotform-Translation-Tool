```python
def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

```python
from collections import OrderedDict
from operator import getitem
```

```python
english_sentences = []
index = 0
for id in english_ids:

  questions = jotformAPI.get_form_questions(id)
  questions = OrderedDict(sorted(questions.items(),  key = lambda x: x[1]['order']))
  for i in questions:
      try:
        text = questions[i]['text']
        clean_text = cleanhtml(text)
        if(clean_text == "Page Break"):
          continue
        english_sentences.append(clean_text)
        print(index)
      except KeyError:
        pass
  index += 1


    
```

```python
lang2_sentences = []
index = 0
for id in lang2_ids:

  questions = jotformAPI.get_form_questions(id)
  questions = OrderedDict(sorted(questions.items(),  key = lambda x: x[1]['order']))
  for i in questions:
      try:
        text = questions[i]['text']
        clean_text = cleanhtml(text)
        if(clean_text == "Page Break"):
          continue
        lang2_sentences.append(clean_text)
        print(index)
      except KeyError:
        pass
  index +=1
```

```python
t = []
e = []
index = 0
for row in dataset:
  lang2_list = nltk.tokenize.sent_tokenize(row[0])
  en_list = nltk.tokenize.sent_tokenize(row[1])

  if(len(lang2_list)!= len(en_list)):
    continue

  for sentence in lang2_list:
    t.append(sentence)

  for sentence in en_list:
    e.append(sentence)
```

```python
print(len(e))
print(len(t))
```

```python
dataset = np.vstack((t,e))
dataset = dataset.T
```

```python
problem_rows = []
index = 0
for row in dataset:
  english = row[0]
  lang2 = row[1]

  if "\n" in english or "\n" in lang2:
    problem_rows.append(index)

  elif len(english) < 2 or len(lang2) < 2:
    problem_rows.append(index)

  elif not any(c.isalpha() for c in english):
    problem_rows.append(index)

  elif not any(c.isalpha() for c in lang2):
    problem_rows.append(index)
  index +=1
```

```python
dataset = np.delete(dataset, problem_rows, axis=0)
print(dataset.shape)
```

```python
# function to read raw text file
def read_text(filename):
    # open the file
    file = open(filename, mode='rt', encoding='utf-8')
    # read all text
    text = file.read()
    file.close()
    return text

# split a text into sentences
def to_lines(text):
    sents = text.strip().split('\n')
    sents = [i.split('\t') for i in sents]
    return sents
```
